```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'ed_comparison.csv'
data = pd.read_csv(file_path)

# Extract the relevant columns for Euclidean distances
# Adjust the column names based on your CSV structure
yolov8_distances = data['YOLOv8'].dropna().values
mediapipe_distances = data['MediaPipe'].dropna().values
openpose_distances = data['OpenPose'].dropna().values

# Create a range for the x-axis (number of data points)
x = np.arange(len(yolov8_distances))

# Plotting
plt.figure(figsize=(35, 15))
plt.plot(x, yolov8_distances, label='YOLOv8-Poses', marker='o')
plt.plot(x, mediapipe_distances, label='MediaPipe', marker='s')
plt.plot(x, openpose_distances, label='OpenPose', marker='^')

# Adding titles and labels
plt.title('Comparison of Euclidean Distances for Different Frameworks')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
plt.legend()
plt.grid()

# Show the plot
plt.show()
```
